```python
import pandas as pd
```

```python
dfmay = pd.read_csv('Resources/201805-citibike-tripdata.csv', engine='python')
```

```python
dfjune = pd.read_csv('Resources/201806-citibike-tripdata.csv', engine='python')
dfjuly = pd.read_csv('Resources/201807-citibike-tripdata.csv', engine='python')
dfaug = pd.read_csv('Resources/201808-citibike-tripdata.csv', engine='python')
dfsept = pd.read_csv('Resources/201809-citibike-tripdata.csv', engine='python')
dfoct = pd.read_csv('Resources/201810-citibike-tripdata.csv', engine='python')
dfnov = pd.read_csv('Resources/201811-citibike-tripdata.csv', engine='python')
dfdec = pd.read_csv('Resources/201812-citibike-tripdata.csv', engine='python')
dfjan = pd.read_csv('Resources/201901-citibike-tripdata.csv', engine='python')
dffeb = pd.read_csv('Resources/201902-citibike-tripdata.csv', engine='python')
dfmar = pd.read_csv('Resources/201903-citibike-tripdata.csv', engine='python')
dfapr = pd.read_csv('Resources/201904-citibike-tripdata.csv', engine='python')
```

```python
dfyear = [dfmay, dfjune, dfjuly, dfaug, dfsept, dfoct, dfnov, dfdec, dfjan, dffeb, dfmar, dfapr]
for month in dfyear:
    month.rename(columns = {'Trip Duration': 'tripduration', 'Start Time': 'starttime', 
                      'Stop Time': 'stoptime', 'Gender':'gender', "User Type": "usertype",
                      'Bike ID': 'bikeid', 'Birth Year': 'birth year', 'Start Station ID': 'start station id',
                      'Start Station Name': 'start station name', 'Start Station Latitude': 'start station latitude',
                      'Start Station Longitude': 'start station longitude', 'End Station ID': 'end station id', 
                      'End Station Name': 'end station name', 'End Station Latitude': 'end station latitude',
                      'End Station Longitude': 'end station longitude'}, inplace = True)
dfyear[5].head()
```

```python
dfcitibike2019 = pd.concat([dfmay, dfjune, dfjuly, dfaug, dfsept, dfoct, dfnov, dfdec, dfjan, dffeb, dfmar, dfapr])
```

```python
dfcitibike2019 = dfcitibike2019.reset_index(drop=True)
dfcitibike2019.head()
```
